```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df['Category'].value_counts()
```

```python
df.replace({'spam':0,'ham':1},inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
X=df['Message']
y=df['Category']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.8,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#feature extraction
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
#model training
model=LogisticRegression()
```

```python
model.fit(X_train_features,y_train)
```

```python
#model eval
#prediction on training data
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
```

```python
print("the training accuracy is:",accuracy_on_training_data)
```

```python
preciction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(y_test,preciction_on_test_data)
```

```python
print("the accuracy on the test data is:",accuracy_on_test_data)
```

```python
input_mail = ["sale! sale! sale! buy 1 get 1 free"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
